```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
from pathlib import Path
```

```python
# reading images
filepath = str(Path("../imgs/beach.jpg"))
gray_scale_img = cv2.imread(filepath, 0)
rgb_scale_img = cv2.imread(filepath)
type(gray_scale_img)
```

```python
cv2.imshow("gray_scale_img", gray_scale_img)
cv2.imshow("rgb_scale_img", rgb_scale_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite("gray_scale_img.jpg", gray_scale_img)
```

```python
plt.imshow(rgb_scale_img) # cv2 reads in BGR format and matplotlib uses RGB
```

## Draw the shapes

```python
rectangle_draw = cv2.rectangle(rgb_scale_img, (384,0),(510,128),(0,150,255),3)
cv2.imshow("rectangle_draw", rectangle_draw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
font = cv2.FONT_ITALIC
predicted_class = "Rectangle" #<< from your model
rectangle_draw = cv2.rectangle(rgb_scale_img, (384,0),(510,128),(0,150,255),3)
cv2.putText(rectangle_draw, predicted_class, (384 + 3,0 + 30), font, 1, (255, 0, 0), 2)
cv2.imshow("rectangle_draw", rectangle_draw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
font = cv2.FONT_ITALIC

circle_draw = cv2.circle(rgb_scale_img, (500, 500),100,(0,150,255),3)
cv2.imshow("circle_draw", circle_draw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
